```python
mean_values = train.mean(axis=0)
train.fillna(mean_values, inplace=True)
train.head
```

```python
x_cols = [col for col in train.columns if col not in ['id', 'timestamp', 'y']]

labels = []
values = []

for col in x_cols:
    labels.append(col)
    value = np.corrcoef(train[col].values, train.y.values)[0, 1]
    values.append(value)
```

```python
sns.set(style="whitegrid")
f, ax = plt.subplots(figsize=(10, 50))
corrs = pd.DataFrame(data={"name": labels, "corr": values})
corrs['corr'] = pd.to_numeric(corrs['corr'])
sns.set_color_codes("pastel")
sns.barplot(x="corr", y="name", data=corrs)
plt.show()
```

```python
cols_to_use = ['technical_30', 'technical_20', 'fundamental_11', 'technical_19']

temp_train = train[cols_to_use]
corrmat = temp_train.corr(method="spearman")
f, ax = plt.subplots(figsize=(8, 8))

sns.heatmap(corrmat, vmax=.8, square=True)
plt.show()
```
