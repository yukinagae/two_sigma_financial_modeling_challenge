```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

pd.set_option('display.max_columns', 120)
```

```python
with pd.HDFStore("./input/train.h5", "r") as train:
    df = train["train"]
```

```python
df.head()
```

```python
df.describe()
```

```python
labels = []
values = []
for col in df.columns:
    labels.append(col)
    values.append(df[col].isnull().sum())
    print(col, values[-1])
```

```python
sns.set(style="whitegrid")
f, ax = plt.subplots(figsize=(10, 50))
counts = pd.DataFrame(data={"name": labels, "count": values})
counts['count'] = pd.to_numeric(counts['count'])
sns.set_color_codes("pastel")
sns.barplot(x="count", y="name", data=counts)
plt.show()
```
